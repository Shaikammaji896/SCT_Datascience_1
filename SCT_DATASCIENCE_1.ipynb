```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('world_population.csv')

# Display dataset
print(df.head())

# Histogram
plt.figure(figsize=(10,6))

plt.hist(df['2022 Population'], bins=15,
         color='skyblue',
         edgecolor='black')

plt.title('Distribution of World Population')
plt.xlabel('Population')
plt.ylabel('Frequency')

plt.show()

# Top 10 countries
top10 = df.sort_values(by='2022 Population',
                       ascending=False).head(10)

# Bar chart
plt.figure(figsize=(12,6))

plt.bar(top10['Country/Territory'],
        top10['2022 Population'],
        color='orange')

plt.title('Top 10 Most Populated Countries (2022)')
plt.xlabel('Country')
plt.ylabel('Population')

plt.xticks(rotation=45)

plt.show()
```
